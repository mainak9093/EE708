```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, mean_squared_error
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Corrected URLs with quotes
train_data = pd.read_csv('https://github.com/mainak9093/EE708/raw/main/A4_train.csv')
test_data = pd.read_csv('https://github.com/mainak9093/EE708/raw/main/A4_test.csv')


print("Training data shape:", train_data.shape)
print("Test data shape:", test_data.shape)

X_train = train_data[['Feature1', 'Feature2']]
y_train = train_data['Label']
X_test = test_data[['Feature1', 'Feature2']]
y_test = test_data['Label']


full_tree = DecisionTreeClassifier(criterion='gini', random_state=42)
full_tree.fit(X_train, y_train)

y_train_pred_full = full_tree.predict(X_train)
y_test_pred_full = full_tree.predict(X_test)


train_accuracy_full = accuracy_score(y_train, y_train_pred_full)
test_accuracy_full = accuracy_score(y_test, y_test_pred_full)
sse_full = mean_squared_error(y_test, y_test_pred_full) * len(y_test)

print(f"\nFull Tree Results:")
print(f"Tree Depth: {full_tree.get_depth()}")
print(f"Number of Leaves: {full_tree.get_n_leaves()}")
print(f"Training Accuracy: {train_accuracy_full:.4f}")
print(f"Test Accuracy: {test_accuracy_full:.4f}")
print(f"SSE on Test Set: {sse_full:.4f}")


plt.figure(figsize=(20, 12))
plot_tree(full_tree,
          feature_names=['Feature1', 'Feature2'],
          class_names=['0', '1'],
          filled=True,
          fontsize=8)
plt.title("Fully Grown Decision Tree (Depth 6)", fontsize=16)
plt.show()


max_depth = full_tree.get_depth()
depths = range(1, max_depth + 1)
sse_values = []
sse_changes = []
test_accuracies = []

print("\nDepth\tSSE\tSSE Change\tTest Accuracy")
print("-" * 40)

for depth in depths:
    tree = DecisionTreeClassifier(criterion='gini', max_depth=depth, random_state=42)
    tree.fit(X_train, y_train)

    y_test_pred = tree.predict(X_test)


    sse = mean_squared_error(y_test, y_test_pred) * len(y_test)
    accuracy = accuracy_score(y_test, y_test_pred)

    sse_values.append(sse)
    test_accuracies.append(accuracy)


    if len(sse_values) > 1:
        sse_change = sse_values[-2] - sse_values[-1]
        sse_changes.append(sse_change)
        print(f"{depth}\t{sse:.1f}\t{sse_change:.1f}\t\t{accuracy:.4f}")
    else:
        sse_changes.append(0)
        print(f"{depth}\t{sse:.1f}\tN/A\t\t{accuracy:.4f}")


plt.figure(figsize=(10, 6))
plt.plot(depths, sse_values, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Tree Depth')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('SSE vs Tree Depth')
plt.grid(True, alpha=0.3)
plt.xticks(depths)


plt.annotate('Significant SSE reduction',
             xy=(3, sse_values[2]),
             xytext=(4, sse_values[2] + 1),
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=12, color='red')
plt.show()


results_df = pd.DataFrame({
    'Depth': depths,
    'SSE': sse_values,
    'SSE_Change': sse_changes,
    'Test_Accuracy': test_accuracies
})

print("\nSSE Analysis Results:")
print(results_df)


print("\nPart C: Optimal Pruning Depth Analysis")
print("=" * 50)


print("SSE Changes by depth:")
for i, depth in enumerate(depths):
    if i == 0:
        print(f"Depth {depth}: Initial SSE = {sse_values[i]}")
    else:
        change = sse_values[i-1] - sse_values[i]
        print(f"Depth {depth}: SSE change = {change} (from {sse_values[i-1]} to {sse_values[i]})")


optimal_depth = 3

print(f"\nOptimal Pruning Depth: {optimal_depth}")
print(f"Reasoning: Depth {optimal_depth} provides the main SSE reduction (from 9.0 to 4.0),")
print(f"while deeper trees (depths 4-6) show no further improvement.")

optimal_tree = DecisionTreeClassifier(criterion='gini', max_depth=optimal_depth, random_state=42)
optimal_tree.fit(X_train, y_train)

y_train_pred_optimal = optimal_tree.predict(X_train)
y_test_pred_optimal = optimal_tree.predict(X_test)

train_accuracy_optimal = accuracy_score(y_train, y_train_pred_optimal)
test_accuracy_optimal = accuracy_score(y_test, y_test_pred_optimal)
sse_optimal = mean_squared_error(y_test, y_test_pred_optimal) * len(y_test)

print(f"\nOptimal Tree (Depth {optimal_depth}) Results:")
print(f"Tree Depth: {optimal_tree.get_depth()}")
print(f"Number of Leaves: {optimal_tree.get_n_leaves()}")
print(f"Training Accuracy: {train_accuracy_optimal:.4f}")
print(f"Test Accuracy: {test_accuracy_optimal:.4f}")
print(f"SSE on Test Set: {sse_optimal:.4f}")


plt.figure(figsize=(15, 10))
plot_tree(optimal_tree,
          feature_names=['Feature1', 'Feature2'],
          class_names=['0', '1'],
          filled=True,
          fontsize=10)
plt.title("Pruned Decision Tree (Optimal Depth 3)", fontsize=16)
plt.show()



print("\nPerformance Comparison: Full Tree vs Pruned Tree")
print("=" * 55)

comparison_data = {
    'Metric': ['Tree Depth', 'Number of Leaves', 'Training Accuracy', 'Test Accuracy', 'SSE on Test Set'],
    'Full Tree': [full_tree.get_depth(), full_tree.get_n_leaves(), f"{train_accuracy_full:.4f}",
                  f"{test_accuracy_full:.4f}", f"{sse_full:.4f}"],
    'Pruned Tree (Depth 3)': [optimal_tree.get_depth(), optimal_tree.get_n_leaves(),
                               f"{train_accuracy_optimal:.4f}", f"{test_accuracy_optimal:.4f}",
                               f"{sse_optimal:.4f}"]
}

comparison_df = pd.DataFrame(comparison_data)
print(comparison_df.to_string(index=False))

print("\n\nKey Findings:")
print("-" * 20)
print("1. Equivalent Test Performance: Both trees achieve identical test accuracy (95.56%) and SSE (4.0)")
print("2. Reduced Overfitting: Pruned tree shows balanced training accuracy (97.14%) vs perfect (100%) for full tree")
print("3. Structural Simplification: Pruned tree reduces complexity by 50% (8 vs 16 leaves)")
print("4. Optimal Complexity: Depth 3 captures essential decision boundaries without unnecessary complexity")

results_df.to_csv('sse_analysis_results.csv', index=False)
comparison_df.to_csv('tree_comparison_results.csv', index=False)

print(f"\nResults saved to CSV files:")
print("- sse_analysis_results.csv")
print("- tree_comparison_results.csv")

print(f"\nAnalysis completed successfully!")
print(f"Optimal pruning depth: {optimal_depth}")
```
